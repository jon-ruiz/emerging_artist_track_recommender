```python
recommender = pd.DataFrame(index = emerging_arrays.keys(), columns=established_arrays.keys())
```

```python
from tqdm import tqdm 
for k, v in established_arrays.items():
    for i, j in tqdm(emerging_arrays.items()):
        try:
            var1 = v.reshape(1, -1)
            var2 = j.reshape(1, -1)
            similarity = cosine_similarity(var1, var2)
            recommender.loc[i,k] = similarity[0][0]
        except:
            recommender.loc[i,k] = 0
```

```python
import random
selection = random.sample(set(recommender.index), 1000)
```

```python
recommender_model = pd.DataFrame(index = selection, columns=["6mFkJmJqdDVQ1REhVfGgd1","1595LW73XBxkRk2ciQOHfr","04ZTP5KsCypmtCmQg5tH9R","37f4ITSlgPX81ad2EvmVQr", "6uAm7pG66O1XesXS9bpHSF","4PpuH4mxL0rD35mOWaLoKS"])
```

```python
recommender_model.columns
```

```python
for i in recommender_model.index:
    array1 = emerging_arrays[i]
    var1 = array1.reshape(1,-1)
    for k in tqdm(recommender_model.columns):
        try:
            array2 = established_arrays[k]
            var2= array2.reshape(1,-1)
            similarity = cosine_similarity(var1, var2)
            recommender_model.loc[i,k] = similarity[0][0]
        except:
            recommender_model.loc[i,k] = 0
```

```python
established = pd.read_csv("/Users/jonruiz/Capstone/Datasets/established_artists_songs.csv")
emerging = pd.read_csv("/Users/jonruiz/Capstone/Datasets/emerging_artists_songs.csv")
```

```python
established.head()
```

```python
established_dict = {}
for i in range(len(established)):
    value = established.track_name[i] + " by " + established.artist[i]
    key = established.track_id[i]
    established_dict[key] = value
```

```python
emerging_dict = {}
for i in range(len(emerging)):
    value = emerging.track_name[i] + " by " + emerging.artist[i]
    key = emerging.track_id[i]
    emerging_dict[key] = value
```

```python
recommender_model.rename(index = emerging_dict, columns= established_dict, inplace= True)
```
